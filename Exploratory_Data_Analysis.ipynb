#**Loading individual cryptocurrency datasets**

```python
df_btc = pd.read_csv('/content/coins/coin_Bitcoin.csv')
df_eth = pd.read_csv('/content/coins/coin_Ethereum.csv')
df_bin = pd.read_csv('/content/coins/coin_BinanceCoin.csv')
df_tet = pd.read_csv('/content/coins/coin_Tether.csv')
df_sol = pd.read_csv('/content/coins/coin_Solana.csv')
df_car = pd.read_csv('/content/coins/coin_Cardano.csv')
df_usd = pd.read_csv('/content/coins/coin_USDCoin.csv')
df_xrp = pd.read_csv('/content/coins/coin_XRP.csv')
df_pol = pd.read_csv('/content/coins/coin_Polkadot.csv')
df_dog = pd.read_csv('/content/coins/coin_Dogecoin.csv')
```

```python
df_btc.head()
```

#**Creating a DataFrame by combining the 'Close' prices of different cryptocurrencies**

```python
df = pd.DataFrame({'BTC': df_btc['Close'],'ETH': df_eth['Close'], 'BIN': df_bin['Close'], 'TET': df_tet['Close'], 'SOL': df_sol['Close'], 'CAR': df_car['Close'], 'USD': df_usd['Close'], 'XRP': df_xrp['Close'], 'POL': df_pol['Close'],'DOGE': df_dog['Close']})
df.head()
```

#**Saving the DataFrame to a CSV file named without including the index.**

```python
dataset_name = 'Cryptocurrency.csv'
df.to_csv(dataset_name, index=False)
```

```python
df.describe()
```

#**Plotting the 'Close' prices for each cryptocurrency in the DataFrame.**

```python
import matplotlib.pyplot as plt
plt.style.use('default')
my_crypto = df
plt.figure(figsize = (12.2, 4.5))
for c in my_crypto.columns.values:
   plt.plot(my_crypto[c], label = c)
plt.title('Cryptocurrency Graph')
plt.xlabel('Days')
plt.ylabel(' Crypto Price ($)')
plt.legend(my_crypto.columns.values, loc= 'upper left')
plt.show()
```

#**Scaling the cryptocurrency DataFrame**

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0, 100)) # Creating a MinMaxScaler object with the specified feature range.
scaled = min_max_scaler.fit_transform(df) # Transforming the original DataFrame and storing the scaled values in the 'scaled' variable.
scaled
```

```python
df_scale = pd.DataFrame(scaled, columns = df.columns) # Creating a new DataFrame ('df_scale') with the scaled values and matching column names.
```

```python
dataset_name = 'Scaled_Cryptocurrency.csv'
df_scale.to_csv(dataset_name, index=False)
```

#**Plotting cryptocurrency prices using the Min-Max scaling technique.**


```python
my_crypto = df_scale
plt.figure(figsize=(12.4, 4.5))
for c in my_crypto.columns.values:
   plt.plot(my_crypto[c], label=c)
plt.title('Cryptocurrency Scaled Graph')
plt.xlabel('Days')
plt.ylabel('Crypto Scaled Price ($)')
plt.legend(my_crypto.columns.values, loc = 'upper left')
plt.show()
```

#**# Calculate Daily Stock Returns**

```python
DSR = df.pct_change(1)
DSR
```

#**Plot the daily returns for each stock**

```python
plt.figure(figsize=(12, 4.5))
for c in DSR.columns.values:
   plt.plot(DSR.index, DSR[c], label = c, lw = 2, alpha = .7)
plt.title('Daily Simple Returns')
plt.ylabel('Percentage (in decimal form')
plt.xlabel('Days')
plt.legend(DSR.columns.values, loc= 'upper right')
plt.show()
```

#**Cryptocurrency Volatility**

```python
print('The cryptocurrency volatility:')
DSR.std()
```

#**Heatmap of the Correlation of Daily Stock Returns**

```python
import seaborn as sns
plt.subplots(figsize= (11,11))
sns.heatmap(DSR.corr(), annot= True, fmt= '.2%')
```

#**Calculate the Cumulative Product of Daily Stock Returns**

```python
DCSR = (DSR+1).cumprod()
DCSR
```

#**Plot the cumulative simple returns for each stock**

```python
plt.figure(figsize=(12.2, 4.5))
for c in DCSR.columns.values:
  plt.plot(DCSR.index, DCSR[c], lw=2, label= c)
plt.title('Daily Cumulative Simple Return')
plt.xlabel('Days')
plt.ylabel('Growth of $1 investment')
plt.legend(DCSR.columns.values, loc = 'upper left', fontsize = 10)
plt.show()
```
