#**Loading the Datasets**

```python
import pandas as pd

df_btc = pd.read_csv('/content/Bitcoin_11_26_2023-12_26_2023_historical_data_coinmarketcap.csv')
df_eth = pd.read_csv('/content/Ethereum_11_26_2023-12_26_2023_historical_data_coinmarketcap.csv')
df_usdt = pd.read_csv('/content/Tether USDt_11_26_2023-12_26_2023_historical_data_coinmarketcap.csv')
df_bnb = pd.read_csv('/content/BNB_11_26_2023-12_26_2023_historical_data_coinmarketcap.csv')
df_xrp = pd.read_csv('/content/XRP_11_26_2023-12_26_2023_historical_data_coinmarketcap.csv')
```

#**Droping the original concatenated column**

```python
df_btc[['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name', 'open', 'high', 'low', 'close', 'volume', 'marketCap', 'timestamp']] = df_btc['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'].str.split(';', expand=True)
df_eth[['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name', 'open', 'high', 'low', 'close', 'volume', 'marketCap', 'timestamp']] = df_eth['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'].str.split(';', expand=True)
df_usdt[['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name', 'open', 'high', 'low', 'close', 'volume', 'marketCap', 'timestamp']] = df_usdt['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'].str.split(';', expand=True)
df_bnb[['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name', 'open', 'high', 'low', 'close', 'volume', 'marketCap', 'timestamp']] = df_bnb['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'].str.split(';', expand=True)
df_xrp[['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name', 'open', 'high', 'low', 'close', 'volume', 'marketCap', 'timestamp']] = df_xrp['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'].str.split(';', expand=True)

df_btc = df_btc.drop(columns=['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'])
df_eth = df_eth.drop(columns=['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'])
df_usdt = df_usdt.drop(columns=['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'])
df_bnb = df_bnb.drop(columns=['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'])
df_xrp = df_xrp.drop(columns=['timeOpen;timeClose;timeHigh;timeLow;name;open;high;low;close;volume;marketCap;timestamp'])
```

#**Deleting the Columns that are not required for our Tableau Visualization**

```python
columns_to_delete = ['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name']

df_btc = df_btc.drop(columns=columns_to_delete)
df_eth = df_eth.drop(columns=columns_to_delete)
df_usdt = df_usdt.drop(columns=columns_to_delete)
df_bnb = df_bnb.drop(columns=columns_to_delete)
df_xrp = df_xrp.drop(columns=columns_to_delete)
```

#**Update the timestamp column format**

```python
df_btc['timestamp'] = df_btc['timestamp'].str.split('T').str[0]
df_eth['timestamp'] = df_eth['timestamp'].str.split('T').str[0]
df_usdt['timestamp'] = df_usdt['timestamp'].str.split('T').str[0]
df_bnb['timestamp'] = df_bnb['timestamp'].str.split('T').str[0]
df_xrp['timestamp'] = df_xrp['timestamp'].str.split('T').str[0]
```

#**Reordering Columns**

```python
df_btc = df_btc[['timestamp'] + [col for col in df_btc.columns if col != 'timestamp']]
df_eth = df_eth[['timestamp'] + [col for col in df_eth.columns if col != 'timestamp']]
df_usdt = df_usdt[['timestamp'] + [col for col in df_usdt.columns if col != 'timestamp']]
df_bnb = df_bnb[['timestamp'] + [col for col in df_bnb.columns if col != 'timestamp']]
df_xrp = df_xrp[['timestamp'] + [col for col in df_xrp.columns if col != 'timestamp']]
```

#**Add a new column "Coin Name" to each DataFrame**

```python
df_btc['Coin Name'] = 'Bitcoin-BTC'
df_eth['Coin Name'] = 'Ethereum-ETH'
df_usdt['Coin Name'] = 'Tether-USDT'
df_bnb['Coin Name'] = 'Binance Coin-BNB'
df_xrp['Coin Name'] = 'XRP-XRP'
```

#**Concatenate the DataFrames**

```python
df = pd.concat([df_btc, df_eth, df_usdt, df_bnb, df_xrp], ignore_index=True)
```
